```python
name = input("Enter student's name: ")
```

```python
score = float(input("Enter student's score: "))
```

```python
if score >= 90 and score <= 100:
    grade = "A"
```

```python
elif score >= 80:
    grade = "B"
```

```python
elif score >= 70:
    grade = "C"
```

```python
elif score >= 0:
    grade = "F"
```

```python
else:
    grade = "Invalid Score"
```

```python
print(f"\nStudent Name: {name}")
```

```python
print(f"Letter Grade: {grade}")
```
